Filling the missing values

```python
# filling the missing value Two ways
# 1. mean - avarage value nemeric
# 2. mode - repeated value
```

```python
# find item weight mean values
df['Item_Weight'].mean()
```

```python
# filling that in the missing value place
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# filling the outlet_size using mode
mode_of_outlet_size = df.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
print(mode_of_outlet_size)
```

```python
missing_values = df['Outlet_Size'].isnull()
print(missing_values)
```

```python
# replace the missing value use mode - lambda function
df.loc[missing_values, 'Outlet_Size'] = df.loc[missing_values, 'Outlet_Type'].apply(lambda outlet_type: mode_of_outlet_size[outlet_type])
```

Data pre prrocessing

```python
# can not be feet string or text data so need to change numeric

# data cleaing
df['Item_Fat_Content'].value_counts()
```

```python
df.replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'}, inplace=True)
```

```python
df['Item_Fat_Content'].value_counts()
```

```python
# label encoding
# can not be feet string or text data so need to change numeric
encoder = LabelEncoder()
```

```python
df.head()
```

```python
df['Item_Identifier'] = encoder.fit_transform(df['Item_Identifier'])
df['Item_Fat_Content'] = encoder.fit_transform(df['Item_Fat_Content'])
df['Item_Type'] = encoder.fit_transform(df['Item_Type'])
df['Outlet_Identifier'] = encoder.fit_transform(df['Outlet_Identifier'])
df['Outlet_Size'] = encoder.fit_transform(df['Outlet_Size'])
df['Outlet_Location_Type'] = encoder.fit_transform(df['Outlet_Location_Type'])
df['Outlet_Type'] = encoder.fit_transform(df['Outlet_Type'])
```

```python
df.head()
```

spliting feature and target

```python
# remove column axis = 1
# remove row axis = 0
x = df.drop(columns='Item_Outlet_Sales', axis=1) # feature
y = df['Item_Outlet_Sales'] # target
```

```python
print(x)
```

```python
print(y)
```

Spliting data into train and test

```python
# 20% of data test data split type = 2
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

ML Model traing
XGBootRegessor

```python
regressor = XGBRegressor()
```

```python
regressor.fit(x_train, y_train)
```

Evaluation

```python
# prediction on traing data
training_data_prediction = regressor.predict(x_train)
```

```python
# R squared value
r2_train = metrics.r2_score(y_train, training_data_prediction)
```

```python
print('R squared value = ', r2_train)
```

```python
# prediction on test data
test_data_prediction = regressor.predict(x_test)
```

```python
# R squared value
r2_test = metrics.r2_score(y_test, test_data_prediction)
```

```python
print('R squared value = ', r2_test)
```
